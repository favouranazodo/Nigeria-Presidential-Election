```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup as BS
import re
import csv
```

### Electoral Geography

```python
url = "https://www.electoralgeography.com/new/en/countries/n/nigeria/nigeria-presidential-election-2015.html"
html_text = requests.get(url)
soup = BS(html_text.content, "html.parser")
```

```python
entry = soup.find_all("table")
```

```python
table1 = entry[0]
```

```python
body = table1.find_all("tr")
```

```python
head = body[0]
```

```python
body_rows = body[1:]
```

```python
headings = []
for item in head.find_all("td"):
    item = (item.text).replace('\n',' ').rstrip()
    headings.append(item)
print(headings)
```

```python
all_rows = []
for row_num in range(len(body_rows)):
    row = []
    for row_item in body_rows[row_num].find_all("td"):
            aa = re.sub("\n","",row_item.text)
            row.append(aa)
    all_rows.append(row)
```

```python
df = pd.DataFrame(data=all_rows, columns=headings)
df
```
