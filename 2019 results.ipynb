# Wikipedia

```python
wikipedia_url = "https://en.wikipedia.org/wiki/2019_Nigerian_general_election#:~:text=Incumbent%20president%20Muhammadu%20Buhari%20won,by%20over%203%20million%20votes."
wikipedia_html = requests.get(wikipedia_url)
```

```python
wikipedia_soup = BS(wikipedia_html.text, 'html.parser')
wikipedia_table = wikipedia_soup.find_all("table", {"class":"wikitable sortable","style":"text-align:right"})
```

```python
wikipedia_table
```

Change to dataframe

```python
wikipedia_df=pd.read_html(str(wikipedia_table))
wikipedia_df=pd.DataFrame(wikipedia_df[0])
```
